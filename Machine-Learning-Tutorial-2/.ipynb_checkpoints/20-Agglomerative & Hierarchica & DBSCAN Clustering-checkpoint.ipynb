## Agglomerative clustering

```python
import mglearn
```

```python
mglearn.plots.plot_agglomerative_algorithm()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(random_state=42)
```

```python
agg=AgglomerativeClustering(n_clusters=3)
```

```python
k=agg.fit_predict(X)
```

```python
mglearn.discrete_scatter(X[:,0],X[:,1],k)
```

## Hierarchical clustering

```python
mglearn.plots.plot_agglomerative()
```

```python
from scipy.cluster.hierarchy import dendrogram, ward
```

```python
X,y=make_blobs(random_state=0,n_samples=12)
linkange_array=ward(X)
```

```python
dendrogram(linkange_array)
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
X,y=make_blobs(random_state=0,n_samples=12)
dbscan=DBSCAN()
```

```python
kumeler=dbscan.fit_predict(X)
kumeler
```

```python
from sklearn.datasets import make_moons
```

```python
X,y=make_moons(n_samples=200, noise=0.06,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X)
X_scaled=scaler.transform(X)
```

```python
dbscan=DBSCAN()
```

```python
kumeler=dbscan.fit_predict(X_scaled)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=kumeler,cmap=mglearn.cm2,s=60)
```
